```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

servico = Service(ChromeDriverManager().install())
nav = webdriver.Chrome(service=servico)

nav.get("https://www.instagram.com/")
time.sleep(2)
#fazendo login
nav.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[1]/div/label/input').send_keys("@username")
nav.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[2]/div/label/input').send_keys("senha", Keys.ENTER)


#cliando no perfil
time.sleep(3)
try:
    nav.find_element(By.CLASS_NAME, '_6q-tv').click()
except:
    pass
time.sleep(2)

try:
    nav.find_element(By.XPATH, '//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]/div').click()
except:
    pass


time.sleep(3)
nav.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[3]/a/div').click()
time.sleep(1)



#fazendo scroll e procurando o perfil desejado para deixa de seguir
for i in range(10000):
    encontrado = False
    try:
        scroll = i * 100
        nav.execute_script(f"document.getElementsByClassName('isgrP')[0].scroll(0, {scroll})")
        time.sleep(1)
        seguidores = nav.find_elements(By.CLASS_NAME, '_7UhW9')
    except:
        pass
    for seguidor in seguidores:
        #nome do perfil alvo
        if '@username do perfil que quer deixar de seguir' in seguidor.text: 
            try:
                nav.find_element(By.XPATH, '/html/body/div[6]/div/div/div/div[3]/ul/div/li[73]/div/div[2]/button/div').click()
                time.sleep(1)
                nav.find_element(By.CLASS_NAME, 'aOOlW').click()
                encontrado = True
            except:
                pass

    if encontrado == True:
        print("deixou de seguir")
        break     
    


```
